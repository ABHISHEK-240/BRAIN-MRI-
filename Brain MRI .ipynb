```python
import os
import zipfile
import urllib.request

# URL to your dataset (replace with the actual URL if needed)
url = 'https://dicom5c.blob.core.windows.net/public/Data.zip'
output_path = '/content/brain_mri_data.zip'

# Download dataset
urllib.request.urlretrieve(url, output_path)

# Unzip the dataset
with zipfile.ZipFile(output_path, 'r') as zip_ref:
    zip_ref.extractall('/content/brain_mri_data')
```

```python
import cv2
import numpy as np
import os
from sklearn.model_selection import train_test_split

# Function to apply CLAHE
def apply_clahe(image, clip_limit=2.0, grid_size=(8, 8)):
    clahe = cv2.createCLAHE(clipLimit=clip_limit, tileGridSize=grid_size)
    return clahe.apply(image)

# Load the dataset
def load_images_masks(image_dir, mask_dir):
    images, masks = [], []
    for filename in os.listdir(image_dir):
        image_path = os.path.join(image_dir, filename)
        mask_path = os.path.join(mask_dir, filename)
        
        if os.path.exists(mask_path):
            image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
            mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
            image = apply_clahe(image)
            images.append(image)
            masks.append(mask)
    return np.array(images), np.array(masks)

image_dir = 'Data/Data'
mask_dir = 'Data/__MACOSX'

images, masks = load_images_masks(image_dir, mask_dir)

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(images, masks, test_size=0.2, random_state=42)

# Normalize the images
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, concatenate
from tensorflow.keras.models import Model

def nested_unet(input_shape):
    inputs = Input(input_shape)

    # Simple U-Net++ encoder-decoder implementation
    conv1 = Conv2D(64, (3, 3), activation='relu', padding='same')(inputs)
    pool1 = MaxPooling2D((2, 2))(conv1)
    
    conv2 = Conv2D(128, (3, 3), activation='relu', padding='same')(pool1)
    pool2 = MaxPooling2D((2, 2))(conv2)

    up1 = UpSampling2D((2, 2))(pool2)
    conv3 = Conv2D(64, (3, 3), activation='relu', padding='same')(up1)
    
    outputs = Conv2D(1, (1, 1), activation='sigmoid')(conv3)
    return Model(inputs, outputs)

input_shape = (256, 256, 1)
nested_model = nested_unet(input_shape)
```

```python
from tensorflow.keras.layers import Multiply

def attention_block(x, g):
    attn = Conv2D(1, (1, 1), activation='sigmoid')(x)
    return Multiply()([x, attn])

def attention_unet(input_shape):
    inputs = Input(input_shape)
    
    # Encoder
    conv1 = Conv2D(64, (3, 3), activation='relu', padding='same')(inputs)
    pool1 = MaxPooling2D((2, 2))(conv1)

    # Attention mechanism
    attn = attention_block(conv1, pool1)
    up1 = UpSampling2D((2, 2))(attn)
    conv2 = Conv2D(64, (3, 3), activation='relu', padding='same')(up1)

    outputs = Conv2D(1, (1, 1), activation='sigmoid')(conv2)
    return Model(inputs, outputs)

attention_model = attention_unet(input_shape)
```

```python
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import binary_crossentropy

# DICE coefficient function
def dice_coef(y_true, y_pred):
    y_true_f = tf.keras.backend.flatten(y_true)
    y_pred_f = tf.keras.backend.flatten(y_pred)
    intersection = tf.keras.backend.sum(y_true_f * y_pred_f)
    return (2. * intersection + 1) / (tf.keras.backend.sum(y_true_f) + tf.keras.backend.sum(y_pred_f) + 1)

# Compile models
nested_model.compile(optimizer=Adam(), loss=binary_crossentropy, metrics=[dice_coef])
attention_model.compile(optimizer=Adam(), loss=binary_crossentropy, metrics=[dice_coef])

# Train models
nested_model.fit(X_train, y_train, epochs=50, validation_data=(X_test, y_test))
attention_model.fit(X_train, y_train, epochs=50, validation_data=(X_test, y_test))
```

```python
# Install and initialize ngrok
from pyngrok import ngrok
import nest_asyncio
import uvicorn

nest_asyncio.apply()

# Set up FastAPI app
from fastapi import FastAPI, UploadFile, File
from tensorflow.keras.models import load_model
import numpy as np
from PIL import Image

app = FastAPI()

# Load the best model
model = load_model("/content/drive/MyDrive/nested_unet_model.h5", custom_objects={"dice_coef": dice_coef})

@app.post("/predict/")
async def predict(file: UploadFile = File(...)):
    image = Image.open(file.file).convert("L")
    image = np.array(image) / 255.0
    image = np.expand_dims(image, axis=(0, -1))
    prediction = model.predict(image)
    return {"prediction": prediction.tolist()}

# Start the FastAPI server using ngrok
public_url = ngrok.connect(8000)
print(f"Public URL: {public_url}")
uvicorn.run(app, host="0.0.0.0", port=8000)
```
